```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity='all'
import numpy as np
import pandas as pd
pd.set_option('display.max_rows',200)
pd.set_option('display.max_columns',50)
```

```python
df1=pd.read_excel('LPL官网-2022LPL春季赛常规赛.xlsx', sheet_name='队伍榜', header=0)
df1
```

```python
df2=pd.read_excel('蜂鸟电竞-2022LPL春季赛常规赛.xlsx', sheet_name='队伍榜', header=0)
df2
```

```python
df3=df2[['排名','一血率','一塔率','局满率','首大龙率','首小龙率','DPM']]
df3
```

```python
df4=pd.merge(df1,df3,how='inner',on='排名')
df4.drop(labels=['出场次数','胜/负','场均金钱','场均大龙','场均小龙'],axis=1,inplace=True)
df4.set_index('排名',inplace=True)
df4=df4.rename(columns={'DPM':'分均伤害'})
```

```python
df4=(
    df4.assign(总击杀=df4['总击杀(场均)'].str.replace(')','').str.split('(',expand=True)[0])
    .assign(场均击杀=df4['总击杀(场均)'].str.replace(')','').str.split('(',expand=True)[1])
    .assign(总死亡=df4['总死亡(场均)'].str.replace(')','').str.split('(',expand=True)[0])
    .assign(场均死亡=df4['总死亡(场均)'].str.replace(')','').str.split('(',expand=True)[1])
)
df4.drop(labels=['总击杀(场均)','总死亡(场均)'],axis=1,inplace=True)
```

```python
df4[['总击杀','总死亡']]=df4[['总击杀','总死亡']].astype(int)
df4[['场均击杀','场均死亡']]=df4[['场均击杀','场均死亡']].astype(float)
```

```python
df4.columns
```

```python
df4=df4[['战队','胜率','总击杀','场均击杀','总死亡','场均死亡','一血率','一塔率','局满率','场均插眼','场均排眼','首大龙率','首小龙率','分均伤害']]
```

```python
df_1A=df4.copy()
df_1A
```

```python
df4['胜率']=df4['胜率'].str.replace('%','').astype(int).apply(lambda x:x/100)
df4['局满率']=df4['局满率'].str.replace('%','').astype(int).apply(lambda x:x/100)
df4['首大龙率']=df4['首大龙率'].str.replace('%','').astype(int).apply(lambda x:x/100)
df4['首小龙率']=df4['首小龙率'].str.replace('%','').astype(int).apply(lambda x:x/100)
df4['一血率']=df4['一血率'].str.replace('%','').astype(float).apply(lambda x:x/100)
df4['一塔率']=df4['一塔率'].str.replace('%','').astype(float).apply(lambda x:x/100)
```

```python
df_1B=df4
df_1B.dtypes
df_1B
```

```python
writerA=pd.ExcelWriter('数据分析A.xlsx',mode='w',engine="openpyxl")
writerB=pd.ExcelWriter('数据分析B.xlsx',mode='w',engine="openpyxl")
df_1A.to_excel(writerA,'队伍榜')
df_1B.to_excel(writerB,'队伍榜')
writerA.save()
writerB.save()
```
