```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity='all'
import numpy as np
import pandas as pd
pd.set_option('display.max_rows',200)
pd.set_option('display.max_columns',50)
```

```python
df1=pd.read_excel('LPL官网-2022LPL春季赛常规赛.xlsx', sheet_name='选手榜', header=0)
df1.set_index('排名',inplace=True)
df1
```

```python
df2=pd.read_excel('蜂鸟电竞-2022LPL春季赛常规赛.xlsx', sheet_name='选手榜', header=0)
df2.set_index('排名',inplace=True)
df2
```

```python
df3=df2[['选手','战队']]
df3
```

```python
df4=pd.merge(df1,df3,how='left',left_on='队员',right_on='选手')
df4.drop('选手',axis=1,inplace=True)
df4
```

```python
df4.columns
```

```python
df4=df4[['队员', '战队' , '位置', '出场次数', 'MVP次数', '总击杀(场均)', '总助攻(场均)', '总死亡(场均)', 'KDA',
       '场均金钱', '场均补刀', '场均插眼', '场均排眼', '场均参团率', '场均对位经济差', '伤害占比', '经济占比']]
```

```python
df4.战队[0]='TES'
df4.战队[54]='TT'
df4.战队[60]='FPX'
df4.战队[72]='LGD'
df4.战队[78]='IG'
df4.战队[82]='LGD'
df4.战队[97]='RA'
```

```python
df4.set_index(['战队','位置'],inplace=True)
df4
```

```python
df4=df4.sort_values(by='战队')
df4
```

```python
df4=df4.drop('KDA',axis=1)
df4
```

```python
df4=(
    df4.assign(总击杀=df4['总击杀(场均)'].str.replace(')','').str.split('(',expand=True)[0])
    .assign(场均击杀=df4['总击杀(场均)'].str.replace(')','').str.split('(',expand=True)[1])
    .assign(总助攻=df4['总助攻(场均)'].str.replace(')','').str.split('(',expand=True)[0])
    .assign(场均助攻=df4['总助攻(场均)'].str.replace(')','').str.split('(',expand=True)[1])
    .assign(总死亡=df4['总死亡(场均)'].str.replace(')','').str.split('(',expand=True)[0])
    .assign(场均死亡=df4['总死亡(场均)'].str.replace(')','').str.split('(',expand=True)[1])
)
```

```python
df4.drop(labels=['总击杀(场均)','总助攻(场均)','总死亡(场均)'],axis=1,inplace=True)
```

```python
df4.dtypes
df4[['总击杀','总助攻','总死亡']]=df4[['总击杀','总助攻','总死亡']].astype(int)
df4[['场均击杀','场均助攻','场均死亡']]=df4[['场均击杀','场均助攻','场均死亡']].astype(float)
df4.dtypes
```

```python
df4=df4[['队员','出场次数','MVP次数','总击杀','场均击杀','总助攻','场均助攻','总死亡','场均死亡','场均金钱','场均补刀','场均插眼','场均排眼','伤害占比','经济占比','场均参团率','场均对位经济差']]
```

```python
df_1A=df4.copy()
df_1A.dtypes
df_1A
```

```python
df4['伤害占比']=df4['伤害占比'].str.replace('%','').astype(float).apply(lambda x:x/100)
df4['经济占比']=df4['经济占比'].str.replace('%','').astype(int).apply(lambda x:x/100)
df4['场均参团率']=df4['场均参团率'].str.replace('%','').astype(float).apply(lambda x:x/100)
```

```python
df_1B=df4
df_1B.dtypes
df_1B
```

```python
writerA=pd.ExcelWriter('数据分析A.xlsx',mode='a',engine="openpyxl")
writerB=pd.ExcelWriter('数据分析B.xlsx',mode='a',engine="openpyxl")
df_1A.to_excel(writerA,'选手榜')
df_1B.to_excel(writerB,'选手榜')
writerA.save()
writerB.save()
```
