```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity='all'
import numpy as np
import pandas as pd
pd.set_option('display.max_rows',200)
pd.set_option('display.max_columns',50)
```

```python
df1=pd.read_excel('LPL官网-2022LPL春季赛常规赛.xlsx', sheet_name='英雄榜', header=0)
df1.set_index('排名',inplace=True)
df1.drop('常用队员',axis=1,inplace=True)
df1
```

```python
df1=(
    df1.assign(总击杀=df1['总击杀（场均）'].str.replace(')','').str.split('(',expand=True)[0])
    .assign(场均击杀=df1['总击杀（场均）'].str.replace(')','').str.split('(',expand=True)[1])
    .assign(总助攻=df1['总助攻（场均）'].str.replace(')','').str.split('(',expand=True)[0])
    .assign(场均助攻=df1['总助攻（场均）'].str.replace(')','').str.split('(',expand=True)[1])
    .assign(总死亡=df1['总死亡（场均）'].str.replace(')','').str.split('(',expand=True)[0])
    .assign(场均死亡=df1['总死亡（场均）'].str.replace(')','').str.split('(',expand=True)[1])
)
df1.drop(labels=['总击杀（场均）','总助攻（场均）','总死亡（场均）'],axis=1,inplace=True)
```

```python
df1[['总击杀','总助攻','总死亡']]=df1[['总击杀','总助攻','总死亡']].astype(int)
df1[['场均击杀','场均助攻','场均死亡']]=df1[['场均击杀','场均助攻','场均死亡']].astype(float)
```

```python
df1.columns
```

```python
df1=df1[['英雄名','胜率','出场次数','ban比率','pick比率','总击杀','场均击杀','总助攻','场均助攻','总死亡','场均死亡','KDA']]
```

```python
df_1A=df1.copy()
df_1A.dtypes
df_1A
```

```python
df1['胜率']=df1['胜率'].str.replace('%','').astype(int).apply(lambda x:x/100)
df1['ban比率']=df1['ban比率'].str.replace('%','').astype(int).apply(lambda x:x/100)
df1['pick比率']=df1['pick比率'].str.replace('%','').astype(int).apply(lambda x:x/100)
```

```python
df_1B=df1
df_1B.dtypes
df_1B
```

```python
writerA=pd.ExcelWriter('数据清洗A.xlsx',mode='a',engine="openpyxl")
writerB=pd.ExcelWriter('数据清洗B.xlsx',mode='a',engine="openpyxl")
df_1A.to_excel(writerA,'英雄榜')
df_1B.to_excel(writerB,'英雄榜')
writerA.save()
writerB.save()
```
