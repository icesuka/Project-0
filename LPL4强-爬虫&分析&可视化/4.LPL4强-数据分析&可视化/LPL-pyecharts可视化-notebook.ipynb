```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity='all'
import pandas as pd
pd.set_option('display.max_rows',200)
pd.set_option('display.max_columns',50)
from pyecharts import options as opts
from pyecharts.charts import Scatter,Pie,Bar,Boxplot,Line,Grid
from pyecharts.globals import ThemeType
```

```python
df_1=pd.read_excel('数据清洗B.xlsx', sheet_name='队伍榜', header=0)
df_2=pd.read_excel('数据清洗B.xlsx', sheet_name='选手榜', header=0)
df_3=pd.read_excel('数据清洗B.xlsx', sheet_name='英雄榜', header=0)
df_2=df_2.fillna(method='ffill')
df_1.set_index('战队',inplace=True)
df_2.set_index('队员',inplace=True)
df_3.set_index('英雄名',inplace=True)
df_3.drop('排名',axis=1,inplace=True)
```

```python
df_1
df_2
df_3
```

```python
wz=['上单','打野','中单','ADC','辅助']
df=df_2[(df_2['战队']=='V5')|(df_2['战队']=='RNG')|(df_2['战队']=='TES')|(df_2['战队']=='JDG')]
df=df[df['出场次数']>5]
df['位置'] = df['位置'].astype('category')
df['位置'].cat.reorder_categories(wz, inplace=True)
df.sort_values('位置', inplace=True)
#新增kda列
df=df.assign(KDA=(df.场均击杀+df.场均助攻)/df.场均死亡)
df.KDA=df.KDA.round(2)
df=df[['战队','位置','出场次数','MVP次数','KDA','场均插眼','场均排眼','伤害占比', '经济占比', '场均对位经济差','场均击杀','场均助攻','场均死亡','场均金钱','场均补刀','场均参团率']]
```

```python
#选手榜
df.dtypes
df
```

```python
df1=df_1[['一血率','一塔率','局满率','首大龙率','首小龙率','分均伤害']]
df1['一塔率']=df1['一塔率'].round(3)
l1=list(df1.一血率[:10])
l2=list(df1.一血率[-7:])
l3=list(df1.一塔率[:10])
l4=list(df1.一塔率[-7:])
df1_1=list((l1,l2))
df1_2=list((l3,l4))
df1_1.append(df1_2[0])
df1_3=df1_1.pop(1)
df1_2.insert(1,df1_3)
df1_4=df1_2.pop(0)
l5=list(df1.首大龙率[:10])
l6=list(df1.首大龙率[-7:])
l7=list(df1.首小龙率[:10])
l8=list(df1.首小龙率[-7:])
df1_5=list((l5,l6))
df1_6=list((l7,l8))
df1_5.append(df1_6[0])
df1_7=df1_5.pop(1)
df1_6.insert(1,df1_7)
df1_8=df1_6.pop(0)
```

```python
#LPL4强队伍5位置MVP数量/出场数量
a = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.WESTEROS))
    .add_xaxis(df.index.to_list())
    .add_yaxis('MVP次数',df['MVP次数'].to_list(), stack='s',category_gap='50%')
    .add_yaxis('出场次数',df['出场次数'].to_list(), stack='s',category_gap='50%')
    .set_global_opts(title_opts=opts.TitleOpts(title='LPL4强队伍5位置MVP数量/出场数量'),
                     xaxis_opts=opts.AxisOpts(axislabel_opts={'rotate': 45}))
    )
a.render_notebook()
```

```python
#4强20位选手的kda
b = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.ESSOS))
    .add_xaxis(df.index.to_list())
    .add_yaxis('',df['KDA'].to_list())
    .reversal_axis()
    .set_series_opts(label_opts=opts.LabelOpts(position="right"))
    .set_global_opts(title_opts=opts.TitleOpts(title='4强20位选手的kda'))
    )
b.render_notebook()
```

```python
#场均插眼&场均排眼
c = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.CHALK))
    .add_xaxis(df.index.to_list())
    .add_yaxis('场均插眼',df['场均插眼'].to_list(),stack='s',category_gap='50%')
    .add_yaxis('场均排眼',df['场均排眼'].to_list(),stack='s',category_gap='50%')
    .set_global_opts(title_opts=opts.TitleOpts(title='场均插眼&场均排眼'),
                     xaxis_opts=opts.AxisOpts(axislabel_opts={'rotate': 45}))
    )
c.render_notebook()
```

```python
#伤害占比&经济占比 
d = ( 
    Pie(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))
    .add('',[list(z) for z in zip(df[df.战队=='V5'].位置,df[df.战队=='V5'].经济占比)],center=['30%','30%'],radius=['32%','45%'])
    .add('',[list(z) for z in zip(df[df.战队=='V5'].位置,df[df.战队=='V5'].伤害占比)],center=['30%','30%'],radius=['0%','28%'],label_opts=opts.LabelOpts(is_show=False, position='center'))
    .add('',[list(z) for z in zip(df[df.战队=='RNG'].位置,df[df.战队=='RNG'].经济占比)],center=['70%','30%'],radius=['32%','45%'])
    .add('',[list(z) for z in zip(df[df.战队=='RNG'].位置,df[df.战队=='RNG'].伤害占比)],center=['70%','30%'],radius=['0%','28%'],label_opts=opts.LabelOpts(is_show=False, position='center'))
    .add('',[list(z) for z in zip(df[df.战队=='TES'].位置,df[df.战队=='TES'].经济占比)],center=['30%','77%'],radius=['32%','45%'])
    .add('',[list(z) for z in zip(df[df.战队=='TES'].位置,df[df.战队=='TES'].伤害占比)],center=['30%','77%'],radius=['0%','28%'],label_opts=opts.LabelOpts(is_show=False, position='center'))
    .add('',[list(z) for z in zip(df[df.战队=='JDG'].位置,df[df.战队=='JDG'].经济占比)],center=['70%','77%'],radius=['32%','45%'])
    .add('',[list(z) for z in zip(df[df.战队=='JDG'].位置,df[df.战队=='JDG'].伤害占比)],center=['70%','77%'],radius=['0%','28%'],label_opts=opts.LabelOpts(is_show=False, position='center'))
    .set_global_opts(title_opts=opts.TitleOpts(title='伤害占比&经济占比'), legend_opts=opts.LegendOpts(pos_left='right', orient='vertical'))
)
d.render_notebook()
```

```python
#英雄出场次数&ban&pick
bar = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.DARK,
                                animation_opts=opts.AnimationOpts(animation_delay=1000, animation_easing='elasticOut')
                               ))
    .add_xaxis(df2.index.to_list())
    .add_yaxis('登场次数',df2.出场次数.to_list(),category_gap='50%')
    .extend_axis(
        yaxis=opts.AxisOpts(
            name='ban/pick率',
            type_='value',
            min_=0,
            max_=1,
            position='right',
            axisline_opts=opts.AxisLineOpts(linestyle_opts=opts.LineStyleOpts(color="#d14a61")))
    )
    .set_global_opts(
        datazoom_opts=[opts.DataZoomOpts(),opts.DataZoomOpts(type_='inside')],
        xaxis_opts=opts.AxisOpts(axislabel_opts={'rotate':45}),
        yaxis_opts=opts.AxisOpts(
            axistick_opts=opts.AxisTickOpts(is_show=True),
            splitline_opts=opts.SplitLineOpts(is_show=True))))
line = (
    Line()
    .add_xaxis(df2.index.to_list())
    .add_yaxis('ban率',df2.ban比率.to_list(), yaxis_index=1,label_opts=opts.LabelOpts(is_show=False))
    .add_yaxis('pick率',df2.pick比率.to_list(), yaxis_index=1,label_opts=opts.LabelOpts(is_show=False))
    )
bar.overlap(line)
j = Grid(init_opts=opts.InitOpts(theme=ThemeType.SHINE))
j.add(bar,opts.GridOpts(pos_left='5%',pos_right='20%'),is_control_axis_index=True)
j.render_notebook()
```
